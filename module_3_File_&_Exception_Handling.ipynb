File Pointer


*   tell(): to tell where the pointer is



```python
with open("./sample_data/sample2.txt", "r") as file:
  print(file.tell())
  print(file.read())
  print(file.tell())
  # reads from last index
  print(file.read())
```

```python
with open("./sample_data/sample2.txt", "r") as file:
  print(file.tell())
  print(file.read(4))
  print(file.tell())
  print(file.read())
  print(file.tell())
```



*   seek(): to move the pointer




```python
with open("./sample_data/sample2.txt", "r") as file:
  print(file.tell())
  # 0 --> 4
  file.seek(4)
  print(file.tell())
```

```python
with open("./sample_data/sample2.txt", "r") as file:
  print(file.tell())
  print(file.read(5))
  print(file.tell())
  file.seek(0)
  print(file.tell())
  print(file.read(5))
```

Practice 1:


---
Get the text from sample.txt and count the following


*   Number of lines
*   Number of words
*   Number of characters

and save it to the counter_of_string.txt file





```python
with open("./sample_data/sample.txt","r") as file:
  lines = file.readlines()

  # total no of lines
  total_lines = len(lines)

  # total no of words
  no_of_words = list(map(lambda x: len(x.split()),lines))
  total_no_of_words = sum(no_of_words)

  # cleaning
  # Cleaning \n
  string_list = list(map(str.strip,lines))
  # Cleaning whitespaces
  string_list = list(map(lambda x: x.replace(" ",""),string_list))

  # total no of characters
  no_of_characters = list(map(lambda x: len(x),string_list))
  total_no_of_characters = sum(no_of_characters)

with open("./sample_data/counter_of_string.txt","w") as file:
  file.write(f"Total Lines: {total_lines}\n")
  file.write(f"Total No of Words: {total_no_of_words}\n")
  file.write(f"Total No of Characters: {total_no_of_characters}")
```
