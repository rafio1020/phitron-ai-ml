Different type of Argument
1.   Positional Argument








```python
def adder(a,b,c):
  print(f"{a},{b},{c}")

adder(10,20,30)
```

```python
def adder(*args):
  for i in args:
    print(i)

adder(10,20,23,40)
```




2.   Keyword Argument



```python
def adder(a,b,c):
  print(f"{a},{b},{c}")

adder(a=10,c=40,b=12)
```

```python
def adder(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}: {value}")

adder(a=10,b=20,c=40,n_estimator=40)
```

Return type of a function



```python
def adder(a,b):

  sum=a+b
  return sum
print(type(adder(20,29)))
```

```python
def adder(a,b):

  sum=a+b
  return sum
print(type(adder(20.2,29)))
```

Iterator


```python
s = {10,20,30,40}
s_iter = iter(s)
print(type(s_iter))
print(s_iter)
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))
print()
for i in s:
  print(i)
```

Generator

```python
lst = [x for x in range(500)]
def data_loader(chunk_size,lst):
  for i in range(0,len(lst),chunk_size):
    yield lst[i:i+chunk_size]

x=data_loader(10,lst)
print(x)
print(next(x))
print(next(x))
```
